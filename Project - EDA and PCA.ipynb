```python
# 2.2 Z-score Scaling
df_num=df.drop('Names',axis=1)
```

```python
from scipy.stats import zscore
df_scaled=df_num.apply(zscore)
```

```python
df_scaled.head().T
```

```python
# 2.3 comparing correlation and covariance matrix
cor=df_scaled.corr()
cor
```

```python
cov=df_scaled.cov()
cov
```

```python
# 2.4 comparing outliers before and after scaling
```

```python
df.boxplot(figsize=(15,10), rot=90, fontsize=15)
plt.show()
```

```python
df_scaled.boxplot(figsize=(15,10), rot=90, fontsize=15)
plt.show()
```

```python
# 2.5
from sklearn.decomposition import PCA
pca = PCA(n_components=17)
df_pca = pca.fit_transform(df_scaled)
```

```python
df_pca = pd.DataFrame(df_pca,columns=['PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9','PC10','PC11','PC12','PC13','PC14','PC15','PC16','PC17'])
df_pca.head()
```

```python
# 2.6
cov_mat=np.cov(df_scaled.T)
cov_mat
```

```python
eval, evec = np.linalg.eig(cov_mat)
print('Eigen values: \n%s', eval,'\n')
print('Eigen vectors:\n%s', evec)
```

```python
# 2.7
PCA_1=np.round(pca.components_[0],2)
PCA_1
```

```python
df_pca_loading=pd.DataFrame(pca.components_[[0]], columns=df_scaled.columns)
round(df_pca_loading.head(1), 2)
```

```python
#2.8
eval
```

```python
total=eval.sum()
total
```

```python
var_explained=[(i/total)*100 for i in sorted(eval, reverse=True)]
var_explained
```

```python
cum_var=np.cumsum(var_explained)
cum_var
```

```python
pc_comps = ['PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9','PC10','PC11','PC12','PC13','PC14','PC15','PC16','PC17']
temp = pd.DataFrame(pc_comps,columns=['PC'])
```

```python
temp['Variance']=var_explained
temp['Cumulative Variance']=cum_var
```

```python
temp
```

```python
# 2.9
sns.lineplot(x=range(0,17), y=var_explained, marker='o')
plt.xlabel('Number of Principal Components', fontsize=15)
plt.ylabel('Explained Variance',fontsize=15)
plt.title('Screeplot', fontsize=15)
plt.grid()
plt.show()
```
